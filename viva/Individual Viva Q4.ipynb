## Data Consolidation

```python
import os
import glob
import shutil

os.getcwd()
```

```python
os.chdir('./humpty-dumpty-ir-spectra')
```

```python
os.getcwd()
```

```python
#Standardising all filenames from mixed case to lowercase

for file in os.listdir():
    os.rename(file, file.lower())
    
os.listdir()
```

```python
compounds = ['Toluene', '1-Octene', 'Ethanol']

for compound in compounds:
     os.mkdir(compound)
```

```python
#Organising .txt files into respective folders

for file in os.listdir():
    for i in glob.glob('*toluene.txt'):
        shutil.move(i, 'Toluene')
    for i in glob.glob('*1-octene.txt'):
        shutil.move(i, '1-Octene')
    for i in glob.glob('*ethanol.txt'):
        shutil.move(i, 'Ethanol')
```
