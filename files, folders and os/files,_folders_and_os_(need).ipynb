### 2.3 Creating folders

```python
os.mkdir('people')

for person in ['John', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    print(f'Creating {path}')
    os.mkdir(path)
```

### 2.4 Checking for existence

```python
#Using try-except
for person in ['John', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    try:
        os.mkdir(path)
        print(f'Creating{path}')
    except FileExistsError:
        print(f'{path} already exists; skipping creation.')
```

```python
#Using os.path.exists()
for person in ['John', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    if os.path.exists(path):
        print(f'{path} already exists; skipping creation.')
    else:
        os.mkdir(path)
        print(f'Creating {path}')
```

### 2.5 Copying files

```python
for person in ['John', 'Paul', 'Ringo']:
    path_to_destination = os.path.join('people', person)
    shutil.copy('sp2273_logo.png', path_to_destination)
    print(f'Copied file to {path_to_destination}')
```

## Exercise 8 :  Tidying a collaboration

```python
import os
import glob
import shutil

os.getcwd()
```

```python
os.chdir('./os-collaboration-exercise-data')
```

```python
os.getcwd()
```

```python
month = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

for i in glob.glob('*.txt'):
    new_name = i.replace(i[0:3], month[i[0:3]])
    os.rename(i, new_name)
    
os.listdir()
```

```python
def rename_my_file(old_file_name):
    new_name = old_file_name.replace(old_file_name[0:3], month[old_file_name[0:3]])
    os.rename(old_file_name, new_name)
    return f"{old_file_name} ----> {new_name}"
 
old_list = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']    
    
for i in old_list:    
    rename_my_file(i)
    
#Files already renamed from Step 1, this should work in the instance where they weren't renamed
```
